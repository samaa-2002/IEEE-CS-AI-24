```python
#replace each range in age coulmn to its creosspond category ex:0-17 ->child ,18-25 ->teenage, 26-50 -> adult , 51-55+ ->old
```

```python
def categorize_age(age):
    if age == '0-17':
        return 'child'
    elif age == '18-25':
        return 'teenage'
    elif age == '26-50':
        return 'adult'
    elif age == '51-55' or age == '55+':
        return 'old'
    else:
        return 'unknown'

df['age'] = df['age'].apply(lambda x: categorize_age(x))

df
```
